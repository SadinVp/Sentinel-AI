```python
import pandas as pd

data = pd.read_csv("HTTP Dataset.csv")


print(data.columns)
print(data.head())
print(data['label'].value_counts())
```

```python
data['request'] = (
    data['request']
    .astype(str)
    .str.lower()
    .str.strip()
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import joblib

X_text = data['request']
y = data['label']

vectorizer = TfidfVectorizer(
    ngram_range=(1,2),
    max_features=5000
)

X = vectorizer.fit_transform(X_text)

model = LogisticRegression(max_iter=1000)
model.fit(X, y)

joblib.dump(model, "global_model.joblib1")
joblib.dump(vectorizer, "global_vectorizer.joblib1")

print("✅ Global model trained successfully")
```

```python
def test(req):
    v = vectorizer.transform([req.lower()])
    return "BAD" if model.predict(v)[0] == 1 else "GOOD"

print(test("GET /product?id=10"))
print(test("GET /login?user=admin' OR '1'='1"))
```
